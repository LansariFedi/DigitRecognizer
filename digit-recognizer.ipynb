# Dgigit Recognizer
### Loading libraries

```python
import pandas as pd
import tensorflow as tf
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score
```

### Loading Data

```python
data_train = pd.read_csv('/kaggle/input/digit-recognizer/train.csv')
data_comp = pd.read_csv('/kaggle/input/digit-recognizer/test.csv')
data_train_array = np.array(data_train)
data_comp_array = np.array(data_comp)
x_train = data_train_array[0:40000, 1:785]
y_train = data_train_array[0:40000,0]
x_test = data_train_array[40001:42000, 1:785]
y_test = data_train_array[40001:42000,0]
x_comp = data_comp_array[:, 0:784]
print(data_train.shape)
```

### Training the Model

```python
model = Sequential(
    [
    Dense(100,activation='relu',name = 'layer1'),
    Dense(75,activation='relu',name = 'layer2'),
    Dense(50,activation='relu',name = 'layer3'),
    Dense(25,activation='relu',name = 'layer4'),
    Dense(10,activation='linear',name = 'layer5'),
    ]
)
model.compile(
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),  
    optimizer=tf.keras.optimizers.Adam(0.001),
)
model.fit(x_train, y_train, epochs=50, batch_size=32)
```

### Testing the model

```python
#test_accuracy = model.evaluate(x_test, y_test)
#print('Test Loss:', test_loss)
#print('Test Accuracy:', test_accuracy)
y_pred_prob = model.predict(x_test)
y_pred = np.argmax(y_pred_prob, axis=1)
l=0
for i in range (1999):
#print(y_pred_prob[7],"\n")
    if (y_test[i] != y_pred[i]):
        #print(y_pred[i],"\n")
        #print(y_test[i])
        l=l+1
print("Number of errors: ",l)
print("Accuracy:",((1999-l)/1999)*100,"%")
```

### Competition Submission

```python
y_comp_prob = model.predict(x_comp)
y_comp = np.argmax(y_comp_prob, axis=1)
print(y_comp.shape)
line_numbers = np.arange(1, 28001)[:, np.newaxis]
submission = np.hstack((line_numbers, y_comp[:, np.newaxis]))
print(submission.shape)
print(submission)
np.savetxt('submission.csv', submission, delimiter=',', fmt='%s', header='ImageId,Label')
```
